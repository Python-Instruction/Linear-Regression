```
#import the required pakages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
dataset = pd.read_csv('/content/gdrive/My Drive/future/simple_regression/Salary_Data.csv')
```

```
#store the data into X independent and y dependet variable
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,1].values
```

```
#show the data 
print(X)
print(y)
```

```
 #split the data into training and test using cross_validation
 from sklearn.model_selection import train_test_split
 X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=1/3,random_state=0)
```

```
# you can visualized how many data for training 
print(X_train)
print(y_train)
```

```
#data used for test
print(X_test)
print(y_test)
```

```
#train the data using Linear Regression
from sklearn.linear_model import LinearRegression
simplelinearRegression = LinearRegression()
simplelinearRegression.fit(X_train,y_train)
```

```
#Predicting the value of y_test
y_predict=simplelinearRegression.predict(X_test)
```

```
#campare with y_predict and y_test
print(y_predict)
print(y_test)
```

```
#implement the graph
plt.scatter(X_train, y_train,color='red')
plt.plot(X_train,simplelinearRegression.predict(X_train))
plt.show()
```
